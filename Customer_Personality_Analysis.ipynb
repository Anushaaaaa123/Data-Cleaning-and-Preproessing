```python
import pandas as pd
```

```python
df = pd.read_csv('/content/marketing_campaign.csv')
df.head()
```

```python
df.isnull()
```

```python
df.isnull().sum()
```

```python
df['Income'].min()
```

```python
df['Income'] = df['Income'].fillna(df['Income'].min())
```

```python
df.isnull().sum()
```

```python
df.drop_duplicates()
```

```python
df.rename(columns={'Year_Birth':'BirthofYear','Kidhome':'Kids','Teenhome':'Teens','Dt_Customer':'DateofEnroll','MntWines':'Wine','MntFruits':'Fruits','MntMeatProducts':'Meat','MntFishProducts':'Fish','MntSweetProducts':'Sweets','MntGoldProds':'Gold'},inplace=True)
```

```python
df['DateofEnroll'] = pd.to_datetime(df['DateofEnroll'], format='%d-%m-%Y')
```

```python
print(df['Z_CostContact'].nunique())
print(df['Z_Revenue'].nunique())
```

```python
df.drop(columns=['Z_CostContact', 'Z_Revenue'], inplace=True)
```

```python
mnt_cols = ['Wine', 'Fruits', 'Meat', 'Fish', 'Sweets', 'Gold']
df['TotalSpending'] = df[mnt_cols].sum(axis=1)
```

```python
df['TotalChildren'] = df['Kids'] + df['Teens']
```

```python
df['Enrollment_Duration_Days'] = (pd.to_datetime('today') - df['DateofEnroll']).dt.days
```

```python
df['Marital_Status'].value_counts()
```

```python
df['Marital_Status'] = df['Marital_Status'].replace({'Married':'In_relationship', 'Together':'In_relationship', 'Absurd':'Single', 'Widow':'Single', 'YOLO':'Single', 'Divorced':'Single', 'Alone':'Single'})
```
